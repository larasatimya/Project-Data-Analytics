```python
df['Brand'].unique(), df['Rating'].unique(), df['Category'].unique(), df['Usage_Frequency'].unique(), df['Gender_Target'].unique(), df['Price_USD'].unique(), df['Number_of_Reviews'].unique(), df['Product_Name'].unique()
```

```python
# Menampilkan Brand dengan rata-rata rating tertinggi
brand_rating_avg = df.groupby('Brand')['Rating'].mean().sort_values(ascending=False)
print("\nBrand dengan rata-rata rating tertinggi:")
print(brand_rating_avg.head(10))
```

```python
# Menghitung jumlah produk berdasarkan Brand
brand_counts = df['Brand'].value_counts()
print("\nJumlah produk per Brand:")
print(brand_counts.head(10))
```

```python
# Cek jumlah baris yang kosong
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
top_products = df["Product_Name"].value_counts().nlargest(10)
sns.barplot(x=top_products.values, y=top_products.index, palette="magma")
plt.title("10 Produk Kosmetik Paling Banyak Digunakan")
plt.xlabel("Jumlah Pengguna")
plt.ylabel("Nama Produk")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_brands = df["Brand"].value_counts().nlargest(10)
sns.barplot(x=top_brands.values, y=top_brands.index, palette="flare")
plt.title("10 Brand Kosmetik dengan Pengguna Terbanyak")
plt.xlabel("Jumlah Pengguna")
plt.ylabel("Brand")
plt.show()
```

```python

# Plot Brand dengan rating rata-rata tertinggi
plt.figure(figsize=(12,6))
sns.barplot(x=brand_rating_avg.head(10).index, y=brand_rating_avg.head(10).values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Brand dengan Rata-rata Rating Tertinggi")
plt.xlabel("Brand")
plt.ylabel("Rata-rata Rating")
plt.show()
```

```python
brand_gender_reviews = df.groupby(['Brand', 'Gender_Target'])['Number_of_Reviews'].sum().reset_index()

# Mengambil 10 Brand dengan jumlah review terbanyak
top_brands = brand_gender_reviews.groupby('Brand')['Number_of_Reviews'].sum().nlargest(10).index
top_brand_reviews = brand_gender_reviews[brand_gender_reviews['Brand'].isin(top_brands)]

# Visualisasi menggunakan seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=top_brand_reviews, x='Brand', y='Number_of_Reviews', hue='Gender_Target', palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Brand dengan Jumlah Review Terbanyak Berdasarkan Gender")
plt.xlabel("Brand")
plt.ylabel("Jumlah Review")
plt.legend(title="Gender_Target")
plt.show()
```

```python
# Histogram distribusi rating
plt.figure(figsize=(8, 5))
sns.histplot(df['Rating'], bins=10, kde=True, color="blue")
plt.title("Distribusi Rating Produk")
plt.xlabel("Rating")
plt.ylabel("Frekuensi")
plt.show()

# Boxplot rating berdasarkan gender
plt.figure(figsize=(8, 5))
sns.boxplot(x='Gender_Target', y='Rating', data=df, palette="coolwarm") # Changed x='Gender' to x='Gender_Target'
plt.title("Distribusi Rating Berdasarkan Gender")
plt.show()
```

```python
# Produk dengan jumlah review terbanyak
top_reviewed_products = df[['Product_Name', 'Number_of_Reviews']].sort_values(by='Number_of_Reviews', ascending=False).head(10)
print(top_reviewed_products)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_reviewed_products, x='Product_Name', y='Number_of_Reviews', palette="rocket")
plt.xticks(rotation=45)
plt.title("Top 10 Produk dengan Review Terbanyak")
plt.xlabel("Nama Produk")
plt.ylabel("Jumlah Review")
plt.show()
```

```python
# Histogram distribusi harga
plt.figure(figsize=(8, 5))
sns.histplot(df['Price_USD'], bins=30, kde=True, color="green")
plt.title("Distribusi Harga Produk")
plt.xlabel("Harga")
plt.ylabel("Frekuensi")
plt.show()

# Scatter plot hubungan antara harga dan rating
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Price_USD'], y=df['Rating'], alpha=0.5)
plt.title("Hubungan antara Harga dan Rating")
plt.xlabel("Harga")
plt.ylabel("Rating")
plt.show()
```

```python
# Menentukan threshold rating untuk review positif dan negatif
df['Review Type'] = df['Rating'].apply(lambda x: 'Positive' if x >= 4 else 'Negative')

# Visualisasi perbandingan review positif vs negatif
plt.figure(figsize=(6, 6))
df['Review Type'].value_counts().plot.pie(autopct='%1.1f%%', colors=['blue', 'red'])
plt.title("Distribusi Review Positif vs Negatif")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Price_USD"], bins=30, kde=True, color="blue")
plt.axvline(50, color='red', linestyle='dashed', linewidth=1, label="$50")
plt.axvline(100, color='red', linestyle='dashed', linewidth=1, label="$100")
plt.axvline(200, color='green', linestyle='dashed', linewidth=1, label="$200")
plt.title("Distribusi Harga Produk Kosmetik")
plt.xlabel("Harga (USD)")
plt.ylabel("Jumlah Produk")
plt.legend()
plt.show()
```

```python
# Distribusi berdasarkan jenis kulit
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Skin_Type", order=df["Skin_Type"].value_counts().index, palette="viridis")
plt.title("Distribusi Produk Berdasarkan Jenis Kulit")
plt.xlabel("Jenis Kulit")
plt.ylabel("Jumlah Produk")
plt.xticks(rotation=45)
plt.show()

# Distribusi berdasarkan gender target
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Gender_Target", order=df["Gender_Target"].value_counts().index, palette="coolwarm")
plt.title("Distribusi Produk Berdasarkan Gender Target")
plt.xlabel("Gender Target")
plt.ylabel("Jumlah Produk")
plt.show()
```

```python
# Tren penggunaan produk berdasarkan jenis kulit dan gender
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Skin_Type", hue="Gender_Target", palette="Set2", order=df["Skin_Type"].value_counts().index)
plt.title("Tren Penggunaan Produk Berdasarkan Jenis Kulit dan Gender")
plt.xlabel("Jenis Kulit")
plt.ylabel("Jumlah Produk")
plt.legend(title="Gender Target")
plt.xticks(rotation=45)
plt.show()
```
