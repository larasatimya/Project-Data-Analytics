```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/drive/MyDrive/diamonds.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
# Scatter plot hubungan antara harga dan Karat
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['carat'], y=df['price'], alpha=0.5)
plt.title("Hubungan antara Harga dan Karat")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()
```

```python
# Scatter plot with alpha and edgecolor versi ukuran kecil
plt.figure(figsize=(8, 6))
plt.scatter(df["carat"], df["price"], alpha=0.3, edgecolor="blue", linewidth=0.5)

# Labels and title
plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Scatter Plot dari Carat vs Price dengan Alpha and Edgecolor")

# Show plot
plt.show()
```

```python
# Scatter plot with alpha and edgecolor versi ukuran besar
plt.figure(figsize=(12, 8))
plt.scatter(df["carat"], df["price"], alpha=0.3, edgecolor="blue", linewidth=0.5)

# Labels and title
plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Scatter Plot dari Carat vs Price dengan Alpha and Edgecolor")

# Show plot
plt.show()
```

```python
# Histogram distribusi harga
plt.figure(figsize=(8, 5))
sns.histplot(df['price'], bins=30, kde=False, color="green")
plt.title("Distribusi Harga Produk")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df['cut'], palette="Set1")

# label dan judul
plt.xlabel("Cut Type")
plt.ylabel("Count")
plt.title("Count Plot of Diamond Cut Types")

# Rotasi label x
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplot Harga berdasarkan potongan
plt.figure(figsize=(8, 5))
sns.boxplot(x='cut', y='price', data=df, palette="Set1")
plt.title("Distribusi Harga Berdasarkan Potongan")
plt.xlabel("Cut")
plt.ylabel("Price")
plt.show()
```

```python
order = ['Fair', 'Ideal', 'Good', 'Very Good', 'Premium']
plt.figure(figsize=(8, 5))
sns.boxplot(x='cut', y='price', data=df, order=order, palette="flare")
plt.title("Distribusi Harga Berdasarkan Potongan")
plt.xlabel("Cut")
plt.ylabel("Price")
plt.show()
```
